## Sets
Python Set is the unordered collection of the data type. It is iterable, mutable(can modify after creation), and has unique elements.
In set, the order of the elements is undefined; it may return the changed sequence of the element.The set is created by using a built-in function set(), or a sequence of elements is passed in the curly braces {} and separated by the comma.

````python
Engineering={'Math','Physics','Chemistry','Ethics','Programming'} #What does engineering require?
Egemen={'Literature','Art','Ethics','Math','Programming'} #What does Egemen like?
Sofia={'Physics', 'Chemistry'}
print(Engineering)
print(Egemen)
````

````python
#remember all the things you can do with sets??
Engineering.union(Egemen) #what is the union of these sets
Engineering | Egemen #same operation
````

````python
Engineering.intersection(Egemen) #things we have in common!
Engineering & Egemen #same operation
````

````python
#we can keep going
Sofia & Egemen & Engineering #nothing in common!
Sofia & (Egemen | Engineering) #remember logical operators? We can do all of these here#
````

````python
Sofia.issubset(Engineering) #we got our first Boolean!
Egemen.issubset(Engineering) #nope
````

## Dictionaries - A more flexible data storage

Dictionary is an unordered set of a key-value pair of items. It is like an associative array or a hash table where each key stores a specific value.

It is a composition of key : value pairs. Key can hold any primitive data type (specific value), whereas value is an arbitrary Python object.

The items in the dictionary are separated with the comma (,) and enclosed in the curly braces {}.

For string keys or values, use ' ' or " "

````python
#First, start with "" is the same with ""
name1='Egemen'
name2="Egemen"
print(name1,name2)
````

````python
Person={'Name':'Angeli','Age':21,'Occupation':'Engineer'} #talk about how this would be created as a list!
print(type(Person))
Person
````

````python
#Let's see keys and values
Person.keys()
Person.values()
#You can convert them back to a list!
print(list(Person.keys()))
print(list(Person.values()))
````

````python
Person['Name'] #Give me the name of the person
Person['Age'] #Give me the age of the person
Person['Age']=22 #Let's update the age
Person
````

You can even store dictionaries in dictionaries. This operation is called nesting


````python
Person1={'Name':'Angeli','Age':21,'Occupation':'Engineer'}
Person2={'Name':'Olivia','Age':25,'Occupation':'Teacher'}
Person1['Friends']=Person2 #create an empty list
print(Person1)
#Now I can ask  questions about the friend!
Person1['Friends']['Name']

print(f"{Person1['Name']} is friends with {Person1['Friends']['Name']}") #mind the double quotes ""
````

````python
#what if we have more than 1 friend? Shocking, I know
Person3={'Name':'Egemen','Age':'Young','Occupation':'Unknown'}
Person1['Friends']=[Person2,Person3] #set up a list of friends!
#print(Person1)
print(Person1['Friends'][0]['Name'])
print(Person1['Friends'][1]['Name'])
````

## Variable Assignments
You need to be very careful when you use

```
A=B
```
assignments in Python. It may not do what you expect!


````python
A=[1,2,3]
B=A #Let's assign B to be equal to A
print(A) #nothing unexpected happened.
print(B)
````

````python
#Let's change the first element of A to 99
A[0]=99
print(A)
#What do you think print(B) will give?
print(B) #WHATT???
````

Equal sign in python just means that the two variables refer to the same space in the memory. If you want to create a copy of A, you have to specify that

````python
from copy import deepcopy #this is a module, more on these later!
A=[1,2,3]
B=deepcopy(A) #Let's assign B to be a copy of A
A[0]=99
print(A)
print(B) #That works as expected!
````

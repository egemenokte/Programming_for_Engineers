## Quick Introduction to Regression Analysis

Regression analysis is a powerful statistical tool used to understand the relationship between variables. In its simplest form, linear regression attempts to model the relationship between two variables by fitting a linear equation to observed data. One of the most basic forms of linear regression is a simple linear model, which looks like this:

$$ y = x \times a + b$$

Here, $y$ represents the dependent variable we're trying to predict or understand, $x$ is the independent variable that we think has an impact on $y$, $x_1$ is the slope of the line, indicating the effect of $x$ on $y$, and $x_0$ is the intercept, representing the value of $y$ when $x$ is zero.

For example, if we want to analyze the relationship between income and education level, $y$ could represent income, and $x$ could represent the years of education. This model would help us predict how changes in education level ($x$) might affect income ($y$).

To extend the concept further, regression can also accommodate multiple independent variables, leading to what is known as multiple regression. This allows us to model the relationship between a dependent variable and two or more independent variables. The general form of a multiple regression equation is:

$$ y = w_0 + w_1 \times x_1 + w_2 \times x_2 + \ldots + w_n \times x_n $$

In this equation:
- $y$ is the dependent variable,
- $x_1, x_2, \ldots, x_n$ are the independent variables,
- $w_1, w_2, \ldots, w_n$ are the coefficients that represent the weight or influence of the corresponding independent variables on $y$,
- $w_0$ is the intercept term, which represents the value of $y$ when all the independent variables are equal to zero.

This model is highly versatile and can be adapted to include various types of variables and interactions, making it a staple tool in many fields including economics, engineering, and social sciences, where it helps to analyze and predict outcomes based on multiple influencing factors.

## How to find the weights?

## How to Find the Weights?

The Mean Squared Error (MSE) is a common metric used to measure the accuracy of a regression model. It calculates the average of the squares of the errors, i.e., the average squared difference between the estimated values and the actual value. The MSE formula is given by:

$$ \text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$

Where:
- $n$ is the number of observations.
- $y_i$ is the actual value of the target variable.
- $\hat{y}_i$ is the predicted value from the regression model.

The lower the MSE, the better the model's performance, as it indicates smaller errors between predicted and actual values.

This metric provides a simple, clear measure of model performance and is particularly useful when comparing different regression models or tuning models for better accuracy.

## Linear Algebra Reminder!

When dealing with multiple regression, linear algebra provides a robust framework for efficiently calculating the coefficients (weights) of the regression model. The matrix form of the multiple regression equation can be expressed as:

$$ \mathbf{y} = \mathbf{X} \mathbf{w} $$

Here:
- $\mathbf{y}$ is the vector of observed values of the dependent variable,
- $\mathbf{X}$ is the matrix of the independent variables (often augmented with a column of ones to incorporate the intercept $w_0$),
- $\mathbf{w}$ is the vector of weights we want to find.

To find the weights $\mathbf{w}$ that best fit our data, we typically use the method of least squares, which minimizes the sum of the squares of the residuals (the differences between observed and predicted values). The solution for the weights can be found using the normal equation:

$$ \mathbf{w} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y} $$

This equation involves several key operations in linear algebra:
- $\mathbf{X}^T$ is the transpose of the matrix $\mathbf{X}$,
- $\mathbf{X}^T \mathbf{X}$ is a matrix multiplication, resulting in a square matrix that sometimes needs regularization to ensure it's invertible,
- $(\mathbf{X}^T \mathbf{X})^{-1}$ is the inverse of this matrix,
- $\mathbf{X}^T \mathbf{y}$ results in a vector after matrix multiplication with the vector $\mathbf{y}$.

By applying these operations, we can efficiently compute the weights that best describe the relationship between our independent variables and the dependent variable, enabling predictions and insights into the effects of various factors.

Using the iris dataset, let's predict petal_width, given the other features.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = iris[['sepal_length', 'sepal_width', 'petal_length']]
y = iris['petal_width']
X
```

```python
model = LinearRegression()
model.fit(X, y)
predictions = model.predict(X)
mse = mean_squared_error(y, predictions)
print(f"Mean Squared Error: {mse}")
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
# Display the coefficients
print("Model coefficients:")
for feature, coef in zip(X.columns, model.coef_):
    print(f"{feature}: {coef}")
print(f"\nIntercept (Bias): {model.intercept_}")
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(y, predictions)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=1)
plt.xlabel('Actual Petal Width')
plt.ylabel('Predicted Petal Width')
plt.title(f'Actual vs. Predicted Petal Width. MSE = {mse:.2f}')
plt.axis('equal')
```

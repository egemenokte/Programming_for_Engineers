##### Matrix Multiplication

Matrix multiplication involves combining two matrices to produce a new matrix. The compatibility of matrix sizes is crucial for this operation. Here's a visual guide to understanding this concept:

**Matrices Setup:**

1. Matrix $A$ (2x3 matrix):
2. Matrix $B$ (3x2 matrix):

These sizes are compatible because the number of columns in $A$ matches the number of rows in $B$.

$$
A = \left[ \begin{array}{ccc}
 1 & 2 & 3 \\
 4 & 5 & 6 \\
\end{array} \right],
\quad
B = \left[ \begin{array}{cc}
 7 & 8 \\
 9 & 10 \\
 11 & 12 \\
\end{array} \right]
$$

**Matrix Multiplication Process:**

The resulting matrix $C$ (2x2) is computed by multiplying rows of $A$ with columns of $B$:

$$
C = A \times B = \left[ \begin{array}{cc}
 (1*7 + 2*9 + 3*11) & (1*8 + 2*10 + 3*12) \\
 (4*7 + 5*9 + 6*11) & (4*8 + 5*10 + 6*12) \\
\end{array} \right]
$$

$$
C = \left[ \begin{array}{cc}
 58 & 64 \\
 139 & 154 \\
\end{array} \right]
$$

**Visual Representation:**

The element $C_{11}$ is computed as follows:
- Multiply each element of the first row of $A$ by the corresponding element of the first column of $B$ and add them up:
- $1 \cdot 7 + 2 \cdot 9 + 3 \cdot 11 = 58$

This visual and step-by-step process simplifies the understanding of matrix multiplication, making it easier to grasp how each element of the resulting matrix is computed.

**Okay, but why and how?**

let's say you want to calculate how much money you made after a successful week of sales

| Day       | Apples | Bananas | Cherries | Oranges |
|-----------|--------|---------|----------|---------|
| Monday    | 14     | 30      | 22       | 18      |
| Tuesday   | 19     | 24      | 15       | 20      |
| Wednesday | 17     | 21      | 25       | 22      |
| Thursday  | 20     | 18      | 17       | 24      |
| Friday    | 16     | 27      | 20       | 21      |


and say each apples, bananas, cherries and oranges cost 2,4,3,2 dollars respectively. How do we calculate how much we made each day?



The sales data matrix \( S \) (items sold per day) and the price vector \( p \) (price per item) are as follows:

S = \begin{bmatrix}
14 & 30 & 22 & 18 \\
19 & 24 & 15 & 20 \\
17 & 21 & 25 & 22 \\
20 & 18 & 17 & 24 \\
16 & 27 & 20 & 21
\end{bmatrix} p = \begin{bmatrix}
2 \\
4 \\
3 \\
2
\end{bmatrix} \]

To find the total sales per day, we multiply the sales data matrix \( S \) by the price vector \( p \):

\begin{bmatrix}
14 \times 2 + 30 \times 4 + 22 \times 3 + 18 \times 2 \\
19 \times 2 + 24 \times 4 + 15 \times 3 + 20 \times 2 \\
17 \times 2 + 21 \times 4 + 25 \times 3 + 22 \times 2 \\
20 \times 2 + 18 \times 4 + 17 \times 3 + 24 \times 2 \\
16 \times 2 + 27 \times 4 + 20 \times 3 + 21 \times 2
\end{bmatrix} \

This results in a column vector where each element corresponds to the total sales for a given day.


```python
# Given sales data and prices for each fruit
sales_data = np.array([
    [14, 30, 22, 18],  # Monday
    [19, 24, 15, 20],  # Tuesday
    [17, 21, 25, 22],  # Wednesday
    [20, 18, 17, 24],  # Thursday
    [16, 27, 20, 21]   # Friday
])

# Prices for each fruit (in the same order as the sales data columns)
prices = np.array([[2], [4], [3], [2]])  # Apples, Bananas, Cherries, Oranges, Blueberries

print(sales_data)

print(prices)

sales = sales_data @ prices #this is the money we made on monday through friday!
print(sales)
```

##### Solving systems of equations

Another powerful reason to know linear algebra (among many others) is to solve systems of equations. Imagine the following equation

Let's consider a 3x3 system of linear equations as an example to demonstrate how to solve it using NumPy. Here's a sample system:

1. \( 2x + 3y - z = 5 \)
2. \( 4x - y + 5z = 6 \)
3. \( -x + 2y + 2z = 8 \)

This system can be written in matrix form as \( Ax = b \), where \( A \) is the matrix of coefficients, \( x \) is the vector of unknowns, and \( b \) is the constants vector.


### Coefficient Matrix \(A\)

The matrix of coefficients \(A\) for the system of equations is:

$$
A = \begin{bmatrix}
    2 & 3 & -1 \\
    4 & -1 & 5 \\
    -1 & 2 & 2
\end{bmatrix}
$$

### Vector of Unknowns \(x\)

The vector \(x\) represents the unknowns \(x\), \(y\), and \(z\) which we are solving for:

$$
x = \begin{bmatrix}
    x \\
    y \\
    z
\end{bmatrix}
$$

### Constants Vector \(b\)

The constants vector \(b\) corresponding to the right side of the equations:

$$
b = \begin{bmatrix}
    5 \\
    6 \\
    8
\end{bmatrix}
$$

### System of Equations

Given the matrix A and vector b, the system of equations can be represented as:

$$
\begin{bmatrix}
    2 & 3 & -1 \\
    4 & -1 & 5 \\
    -1 & 2 & 2
\end{bmatrix}
\begin{bmatrix}
    x \\
    y \\
    z
\end{bmatrix}
=
\begin{bmatrix}
    5 \\
    6 \\
    8
\end{bmatrix}
$$

To isolate the vector \(x\), we multiply both sides of the equation by the inverse of matrix \(A\) (\(A^{-1}\)):

$$
A^{-1} \cdot \begin{bmatrix}
    2 & 3 & -1 \\
    4 & -1 & 5 \\
    -1 & 2 & 2
\end{bmatrix}
\begin{bmatrix}
    x \\
    y \\
    z
\end{bmatrix}
=
A^{-1} \cdot \begin{bmatrix}
    5 \\
    6 \\
    8
\end{bmatrix}
$$

Simplifying the left side using the property that $A \cdot A^{-1} = I$ (the identity matrix), we get:

$$
\begin{bmatrix}
    x \\
    y \\
    z
\end{bmatrix}
=
A^{-1} \cdot \begin{bmatrix}
    5 \\
    6 \\
    8
\end{bmatrix}
$$

This results in \(x\), \(y\), and \(z\) being explicitly calculated by multiplying the inverse of \(A\) with the constants vector \(b\). This approach shows how matrix operations can be used to find solutions to systems of linear equations effectively.


```python
# Coefficient matrix A
A = np.array([[2, 3, -1],
              [4, -1, 5],
              [-1, 2, 2]])

# Constants vector b
b = np.array([5, 6, 8])

x=np.linalg.inv(A) @ b
print(x) #these are the solutions!

#But I suggest using
print(np.linalg.solve(A,b)) #This is more stable!
```

###### Tension on Cables

We need to solve for the tensions $T_1$ and $T_2$ in two cables holding a 10 kN load in equilibrium. The system can be broken down into horizontal and vertical components.

### Step 1: Force Equilibrium

- Horizontal equilibrium:

$$T_1 \cdot \cos(30^\circ) - T_2 \cdot \cos(45^\circ) = 0$$

- Vertical equilibrium:

$$T_1 \cdot \sin(30^\circ) + T_2 \cdot \sin(45^\circ) = 10$$

### Step 2: Matrix Form

We can write this system as a matrix equation $Ax = b$, where:

$$
A = \begin{bmatrix}
    \cos(30^\circ) & -\cos(45^\circ) \\
    \sin(30^\circ) & \sin(45^\circ)
\end{bmatrix}
, \quad
x = \begin{bmatrix}
    T_1 \\
    T_2
\end{bmatrix}
, \quad
b = \begin{bmatrix}
    0 \\
    10
\end{bmatrix}
$$

### Step 3: Solve the System

To solve for $T_1$ and $T_2$, multiply both sides of $Ax = b$ by $A^{-1}$:

$$x = A^{-1} \cdot b$$

This gives the values of $T_1$ and $T_2$.

image.png

```python
import numpy as np

# Define angles in radians
theta1 = np.radians(30)
theta2 = np.radians(45)

# Coefficient matrix A
A = np.array([[np.cos(theta1), -np.cos(theta2)],
              [np.sin(theta1), np.sin(theta2)]])

# Constants vector b
b = np.array([0, 10])

# Solve for the unknowns (T1, T2)
tensions = np.linalg.solve(A, b)
T1, T2 = tensions

print(f"T1 = {T1:.2f} kN")
print(f"T2 = {T2:.2f} kN")
```

######Example

Let me simplify the example with just two equations focusing on labor hours and cost.

An environmental firm sells two products:
- Product A: Water filtration systems
- Product B: Solar panels

With the following constraints:
1. Each water filtration system requires 4 hours of labor and each solar panel requires 6 hours of labor. They have 180 hours of total labor available.
2. Each water filtration system costs \$100 to produce and each solar panel costs \$200 to produce. They have a budget of $5000.

How many filters and panels should they produce?

Write and then solve this system using linear set of equations



```python
import numpy as np

# Coefficient matrix A
A = np.array([[4, 6],
              [100, 200]])

# Constants vector b
b = np.array([180, 5000])

# Solve for x and y
solution = np.linalg.solve(A, b)

print(f"Water filtration systems (x) = {solution[0]:.1f}")
print(f"Solar panels (y) = {solution[1]:.1f}")
```
